# Analyse: comparison between HL60 cells and  NB4 cells (acute promyelocytic leukemia cell lines)


### HL60 cells
The HL-60 cell line is a human leukemia cell line that has been used for laboratory research on blood cell formation and physiology.  The cell line was derived from a 36-year-old woman who was originally reported to have acute promyelocytic leukemia at the MD Anderson Cancer Center. HL-60 cells predominantly show neutrophilic promyelocytic morphology. HL-60 cells predominantly show neutrophilic promyelocytic morphology. 

### NB4 cells

acute promyelocytic leukemia.established from the bone marrow of a 23-year-old woman with acute promyelocytic leukemia (APL = AML FAB M3) in second relapse in 1989; patented cell line; cells carry the t(15;17) PML-RARA fusion gene. NB4 cells, an acute promyelocytic leukemia cell line, have the t(15;17) translocation and differentiate in response to ATRA, whereas HL-60 cells lack this chromosomal translocation, even after differentiation by ATRA. Die häufigste Ursache für acute promyelocytic leukemia  ist eine t(15;17)-Translokation, die in einer Fusion der beiden für die Transkriptionsfaktoren Promyelocytic Leukemia Protein (PML) und Retinsäurerezeptor-α (RARα) codierenden Gene resultiert.
